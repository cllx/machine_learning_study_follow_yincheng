```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #决策树分类
from sklearn.tree import DecisionTreeRegressor #决策树回归
import matplotlib.pyplot as plt
%matplotlib inline
path = r"./lenses.txt"
data = pd.read_table(path, header=None)
data
```

```python
data[0].unique()
```

```python
data[1].unique()
```

```python
data[3].unique() #看对应的列不重合的数据是哪些
```

```python
dict0 = {"young":0, "pre":1, "presbyopic":2} #这里是手动把每列那些不重复的值映射为数字
data[0] = data[0].map(dict0)
data[0]
```

```python
dict1 = {"myope":0, "hyper":1}
data[1] = data[1].map(dict1)
data[1]
```

```python
dict2 = {"yes":0, "no":1}
data[2] = data[2].map(dict2)
data[2]
```

```python
dict3 = {"reduced":0, "normal":1}
data[3] = data[3].map(dict3)
data[3]
```

```python
dict4 = {"no lenses":0, "soft":1, "hard":2}
data[4] = data[4].map(dict4)
data[4]
```

```python
data
```

```python
x_train, x_test = data.iloc[:-2, :-1], data.iloc[-2:, :-1] #创建决策树分类
y_train, y_test = data[4][:-2], data[4][-2:] #训练数据
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
y_new = dtc.predict(x_test)
y_new
```
