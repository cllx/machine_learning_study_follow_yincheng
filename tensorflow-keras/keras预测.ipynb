```python
import pandas as pd
import numpy as np
data = pd.read_csv(r"student_data.csv")
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def plot_points(data):
    x = np.array(data[["gre","gpa"]])
    y = np.array(data["admit"])
    ok = x[np.argwhere(y==1)]
    no = x[np.argwhere(y==0)]
    plt.scatter([s[0][0] for s in no],[s[0][1] for s in no],s=25,
                color="red",edgecolors="k")
    plt.scatter([s[0][0] for s in ok],[s[0][1] for s in ok],s=25,
                color="blue",edgecolors="k")
    plt.xlabel("GRE")
    plt.xlabel("GPA")
plot_points(data)
plt.show()
```

```python
data_rank1 = data[data["rank"]==1]
data_rank2 = data[data["rank"]==2]
data_rank3 = data[data["rank"]==3]
data_rank4 = data[data["rank"]==4]
plot_points(data_rank1)
plt.title("rank1")
plt.show()
```

```python
plot_points(data_rank2)
plt.title("rank2")
plt.show()
```

```python
plot_points(data_rank3)
plt.title("rank3")
plt.show()
```

```python
one_hot_data = pd.concat([data,pd.get_dummies(data["rank"],prefix="rank")],axis=1)
one_hot_data
```

```python
one_hot_data = one_hot_data.drop("rank",axis=1)
```

```python
one_hot_data
```

```python
process_data = one_hot_data[:] #æ‹·è´å…¨éƒ¨æ•°æ®
process_data["gre"] = process_data["gre"]/800
process_data["gpa"] = process_data["gpa"]/4.0
process_data
```

```python
sample = np.random.choice(process_data.index, #æ‰“ä¹±é¡ºåº
                          size=int(len(process_data)*0.9),
                          replace = False)
sample
```

```python
train_data,test_data = process_data.iloc[sample],process_data.drop(sample)
print("  ",len(train_data),len(test_data))
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
import keras
features = np.array(train_data.drop("admit",axis=1))
target = np.array(keras.utils.to_categorical(train_data["admit"],2))
features
```

```python
target
```

```python
features_test = np.array(test_data.drop("admit",axis=1))
target_test = np.array(keras.utils.to_categorical(test_data["admit"],2))
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense,Dropout,Activation
from keras.optimizers import SGD
from keras.utils import np_utils
model = Sequential()
model.add(Dense(128,activation="relu",input_shape=(6,))) #è¾“å…¥å±‚
model.add(Dropout(0.2))
model.add(Dense(64,activation="relu")) #éšè—å±‚
model.add(Dropout(0.1))
model.add(Dense(2,activation="softmax")) #è¾“å‡ºå±‚
# ç¼–è¯‘æ¨¡åž‹
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=['accuracy'])
model.summary() #é¢„è§ˆæ¨¡åž‹
```

```python
target[0]
```

```python
model.fit(features,target,epochs=200,batch_size=100,verbose=0) #è®­ç»ƒæ•°æ®
```

```python
score1 = model.evaluate(features,target)
score1
```

```python
score2 = model.evaluate(features_test,target_test)
score2
```
