```python
from keras.datasets import mnist
import keras #深度学习框架
import keras.models #模型
from keras.models import Sequential #神经网络
from keras.layers import Dense #处理神经网络层
from keras.utils import np_utils #numpy处理
from keras.layers import Activation #激活
from keras.layers import SimpleRNN #循环神经网络
from keras.optimizers import Adam #优化
import numpy as np

batch_index = 0
batch_size = 50
time_step = 28
input_size=28
cell_size=50
path='./MNIST_data/mnist.npz'
f = np.load(path)
x_train, y_train = f['x_train'], f['y_train'] #训练数据
x_test, y_test = f['x_test'], f['y_test'] #测试数据
x_train = x_train.reshape(-1,28,28)/255 #调整形状，颜色0-255，百分处理
x_test = x_test.reshape(-1,28,28)/255
y_train = np_utils.to_categorical(y_train,num_classes=10) #分类结果10个
y_test = np_utils.to_categorical(y_test,num_classes=10)

# 建立模型
model = Sequential()
model.add(SimpleRNN(batch_input_shape=(batch_size,time_step,input_size),
                    output_dim=50))
model.add(Dense(10)) #输出的有10类
model.add(Activation("softmax")) #选择最优
# 优化系数
adam = Adam(0.001)
model.compile(optimizer=adam,
              loss="categorical_crossentropy", #损失函数最小
              metrics=["accuracy"]) #精准
for step in range(4001):
    x_batch = x_train[batch_index:batch_index+batch_size,:,:]
    y_batch = y_train[batch_index:batch_index+batch_size,:]
    cost = model.train_on_batch(x_batch,y_batch) #训练
    batch_index += batch_size #每次增加50
    batch_index=0 if batch_index>=x_train.shape[0] else batch_index
    if step%100 == 0:
        print(model.evaluate(x_test,y_test,
                             batch_size=50,
                             verbose=False)) #评分数据
```
