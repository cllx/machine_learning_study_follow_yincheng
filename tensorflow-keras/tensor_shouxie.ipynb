```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import matplotlib.pyplot as plt
# 读取文件
mnist = input_data.read_data_sets("./data", one_hot=True)
print(type(mnist)) #训练集合
```

```python
print("  ",mnist.train.num_examples) #训练数量
print("  ",mnist.test.num_examples) #测试数量
```

```python
import numpy as np
train_img = mnist.train.images #训练图片
train_label = mnist.train.labels #训练标签
test_img = mnist.test.images #测试图片
test_label = mnist.test.labels #测试标签
for i in range(10):
    cur_img = np.reshape(train_img[i,:], (28,28)) #取出图片的二维数组
    cur_label = np.argmax(train_label[i,:]) #取出文本
    plt.matshow(cur_img) #显示图片数组
    plt.title(str(cur_label)) #显示文本
    plt.show()
```
