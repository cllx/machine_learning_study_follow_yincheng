```python
# 贝叶斯邮件分类
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import MultinomialNB, GaussianNB #多项式和高斯
from sklearn.feature_extraction.text import TfidfVectorizer #文本处理工具
```

```python
path1 = r"./ham"
path2 = r"./spam"
lable1 = "ham"
lable2 = "spam" #类型划分
```

```python
ham = pd.read_table(path1, header=None)
print(ham.shape)
ham
```

```python
spam = pd.read_table(path2, header=None)
print(spam.shape)
spam
```

```python
alldf = pd.concat([ham, spam]) #把两个合并
alldf
```

```python
x = alldf[1] #表示取第一列
x
```

```python
tf = TfidfVectorizer() #文本矢量处理
x = tf.fit_transform(x)
x.shape #8713 5581条短信不重复所有单词
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test = train_test_split(x, test_size=0.2, random_state=2)
y_train, y_test = train_test_split(alldf[0], test_size=0.2, random_state=2)
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
mnb.score(x_test, y_test)
```

```python
p_data = mnb.predict(x_test) #预测
p_data
```
