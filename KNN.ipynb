```python
import numpy as np
import sklearn.datasets
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap #绘图引用模块
%matplotlib inline
# 获取数据当成训练样本
iris = sklearn.datasets.load_iris() #引入鸢尾花数据集
x = iris.data[:,:2] #取所有的行，取最后一列
y = iris.target
# 设置KNN K=15，计算周围临近15个点
K = 15
# 图片x,y每一步的步长
h = 0.02
cmap_light = ListedColormap(["#FFAAAA", "#AAFFAA", "#AAAAFF"])
cmap_bold = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])
myknn = KNeighborsClassifier(n_neighbors=K) #设置访问周围15个点
myknn.fit(x, y) #训练数据
# 四个数描述图片显示范围
xmin, xmax = x[:,0].min()-1, x[:,0].max()-1
ymin, ymax = x[:,1].min()-1, x[:,1].max()-1
# 生成网格
xx, yy = np.meshgrid(np.arange(xmin, xmax, h),
                     np.arange(ymin, ymax, h))
# 预测
z = myknn.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)

# 显示背景颜色
plt.pcolormesh(xx, yy, z, cmap=cmap_light)
# 显示点的颜色
plt.scatter(x[:,0], x[:,1], c=y, cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("分类")
plt.show()
```

```python
# KNN预测adult数据集的收入
# 下面注释部分之所以在后面会报错，可能原因是自己在adult2.txt中手动添加内容不符合相应的格式，所以下面不能按对应列标签
# 来取列的数据
import numpy as np
import sklearn.datasets
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier #回归
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap #绘图引用模块
%matplotlib inline
path = r"./adult.txt"
data = pd.read_csv(path, header = None)
data.columns = ['age', 'workclass', 'final_weight', 'education', 'education_num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'salary']
# path = r"./adult2.txt"
# data = pd.read_csv(path)
data
```

```python
data.head()
data.shape
data.index
data.columns
X = data[['age','education','occupation','hours_per_week']]
X
Y = data['salary']
```

```python
X["education"].unique().size
X["education"].unique()
X["occupation"].unique()
```

```python
# 对数据进行量化
def edu2float(edu): #任何学历返回一个数字代表
    return np.argwhere(X["education"].unique() == edu)[0,0]
def occupation2float(occupation): #任何职业返回一个数字代表
    return np.argwhere(X["occupation"].unique() == occupation)[0,0]
X["occupation"] = X["occupation"].map(occupation2float)
X["education"] = X["education"].map(edu2float)
```

```python
# X["education"].shape
# X["education"][12888] 看对应行的学历是多少
X_train = X.iloc[:-1000] #训练数据
X_test = X.iloc[-1000:] #测试数据
Y_train = Y.iloc[:-1000] #训练数据
Y_test = Y.iloc[-1000:] #测试数据
knn = KNeighborsClassifier(15)
knn.fit(X_train, Y_train) #建立对象，训练
Ynew = knn.predict(X_test) #用测试数据产生预测结果
result = (Y_test == Ynew) #统计相等个数
accuracy = result.sum()/1000 #统计成功率
print("=======>%f" %(accuracy))
```

```python
# KNN手写(矩阵形式)识别
import numpy as np
import sklearn.datasets
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier #回归
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap #绘图引用模块
from sklearn.externals import joblib
from sklearn.decomposition import PCA #特征处理，降维
import time
%matplotlib inline

path = r""
train_num = 20000 #训练的数据量
test_num = 30000 #测试的数据量为50000-30000
data = pd.read_csv("mnist_train.csv", header=None) #读取数据
print(data)
train_data = data.values[0:train_num, 1:] #抓取训练数据
train_label = data.values[0:train_num, 0] #训练数据的标签
test_data = data.values[train_num:test_num, 1:] #测试的数据
test_label = data.values[train_num:test_num, 0] #测试的标签
start = time.time()
#特征训练，进行数据的降维 whiten=True白化处理，n_components特征
# n_components：这个参数可以帮我们指定希望PCA降维后的特征维度数目。
# 最常用的做法是直接指定降维到的维度数目，此时n_components是一个大于等于1的整数。
# 当然，我们也可以指定主成分的方差和所占的最小比例阈值，让PCA类自己去根据样本特征方差来决定降维到的维度数，
# 此时n_components是一个（0，1]之间的数
pca = PCA(n_components=200, whiten=True) 
train_x = pca.fit_transform(train_data) #数据预处理
test_x = pca.fit_transform(test_data)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_x, train_label) #训练
pre = knn.predict(test_x) #预测
acc = (pre == test_label).sum()/len(test_label) #计算准确率
end = time.time()
print("准确率%f  time%d" %(acc,end-start))
joblib.dump(knn, r"path") #把训练好的模型保存到对应的路径下
# mynewknn = joblib.load(r"path") #加载已训练好的模型
```

```python
# KNN手写(矩阵形式)识别
import numpy as np
import sklearn.datasets
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier #回归
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap #绘图引用模块
from sklearn.externals import joblib
from sklearn.decomposition import PCA #特征处理，降维
import time
%matplotlib inline

path = r""
train_data = pd.read_csv("mnist_train.csv", header=None) #读取数据
test_data = pd.read_csv("mnist_test.csv", header=None) #读取数据
print(train_data)
print(test_data)
train_data_nolabel = train_data.values[:, 1:] #抓取训练数据
train_label = train_data.values[:, 0] #训练数据的标签
test_data_nolabel = test_data.values[:, 1:] #抓取测试数据
test_label = test_data.values[:, 0] #测试的标签
start = time.time()
#特征训练，进行数据的降维 whiten=True白化处理，n_components特征
# n_components：这个参数可以帮我们指定希望PCA降维后的特征维度数目。
# 最常用的做法是直接指定降维到的维度数目，此时n_components是一个大于等于1的整数。
# 当然，我们也可以指定主成分的方差和所占的最小比例阈值，让PCA类自己去根据样本特征方差来决定降维到的维度数，
# 此时n_components是一个（0，1]之间的数
pca = PCA(n_components=700, whiten=True) 
train_x = pca.fit_transform(train_data_nolabel) #数据预处理
test_x = pca.fit_transform(test_data_nolabel)
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(train_x, train_label) #训练
pre = knn.predict(test_x) #预测
acc = (pre == test_label).sum()/len(test_label) #计算准确率
end = time.time()
print("准确率%f  time:%d" %(acc,end-start))
joblib.dump(knn, r"path") #把训练好的模型保存到对应的路径下
# mynewknn = joblib.load(r"path") #加载已训练好的模型
```
