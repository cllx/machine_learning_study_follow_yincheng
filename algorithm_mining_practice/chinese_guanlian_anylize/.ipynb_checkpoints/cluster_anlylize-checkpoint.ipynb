```python
import pandas as pd
from sklearn.cluster import KMeans

datapath = "data.xls"
data = pd.read_excel(datapath)
print(data)
```

```python
typelabel = {u'肝气郁结证型系数': 'A',
             u'热毒蕴结证型系数': 'B',
             u'冲任失调证型系数': 'C',
             u'气血两虚证型系数': 'D',
             u'脾胃虚弱证型系数': 'E',
             u'肝肾阴虚证型系数': 'F'}
k = 4 #需要进行的聚类类别数
keys = list(typelabel.keys()) #多个选项(名称)
result = pd.DataFrame() #最终保存的结果
if __name__ == "__main__":
    for i in range(len(keys)): #每个字段分类
        kmn = KMeans(n_clusters=k, n_jobs=4) #创建聚类
        kmn.fit(data[[keys[i]]].as_matrix()) #训练模型
        r1 = pd.DataFrame(kmn.cluster_centers_, columns=[typelabel[keys[i]]])
        r2 = pd.Series(kmn.labels_).value_counts() #统计次数
        r2 = pd.DataFrame(r2, columns=[typelabel[keys[i]]+"n"])
        r = pd.concat([r1, r2], axis=1).sort_values(typelabel[keys[i]])
        r.index = [1,2,3,4]
        r[typelabel[keys[i]]] = pd.rolling_mean(r[typelabel[keys[i]]], 2) #计算相邻的平均值，然后分隔开
        r[typelabel[keys[i]]][1] = 0.0 #聚类中心改成边界点
        result = result.append(r.T)
#     result = result.sort_values()
    result.to_excel("dataok.xls")
```
