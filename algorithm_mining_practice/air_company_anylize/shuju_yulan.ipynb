```python
# 数据预览
import pandas as pd
datapath = "air_data.csv"
data = pd.read_csv(datapath, encoding="utf-8")
print(data)
```

```python
explore = data.describe(percentiles=[], include="all").T #。T表示转置
print(explore)
explore["null"] = len(data)-explore['count'] #计算非空
explore = explore[["null", "max", "min"]]
explore.to_excel("dataexplore.xls")
```

```python
# 数据清洗
import pandas as pd
datapath = "air_data.csv"
data = pd.read_csv(datapath,encoding="utf-8")
data = data[data["SUM_YR_1"].notnull()*data["SUM_YR_2"].notnull()]#保留两个都不为空的数据
index1 = data["SUM_YR_1"] != 0
index2 = data["SUM_YR_2"] != 0
index3 = (data["SEG_KM_SUM"]==0) & (data["avg_discount"]==0)
data = data[index1 | index2 | index3]
data.to_excel("cleancsv.xls")
```

```python
# 数据标准化
import pandas as pd
inputdatafile = "zscoredata.xls"
outdatafile = "zscoredata_new.xls"
data = pd.read_excel(inputdatafile)
data = (data-data.mean(axis=0))/(data.std(axis=0)) #标准差的转换
data.columns = ["Z"+i for i in data.columns]
print("-----------------------------------------")
print(data)
data.to_excel(outdatafile)
```
