```python
import pandas as pd
from scipy.interpolate import lagrange
inputpath = "missing_data.xls" #输入
outputpath = "missing_data_ok.xls" #输出
data = pd.read_excel(inputpath, header=None)
print(data)
```

```python
# 自定义列向量插值函数
# s为列向量，n为被插值的位置，k为取前后的数据个数，默认为5
def ployinterp_column(s, n, k=5):
    y = s[list(range(n-k, n))+list(range(n+1, n+1+k))] #取数
    y = y[y.notnull()] #剔除空值
    return lagrange(y.index, list(y))(n) #插值后并返回插值结果

for i in data.columns: #处理缺失数据
    for j in range(len(data)):
        if (data[i].isnull())[j]:
            data[i][j] = ployinterp_column(data[i], j)
print("--------------------------------------")
print(data)
data.to_excel(outputpath, header=None, index=None)
```

```python
# 使用深度学习神经网络挖掘电力
import pandas as pd
import random
datafile = "model.xls"
data = pd.read_excel(datafile)
data = data.as_matrix() #表格转换矩阵数组
random.shuffle(data) #打乱顺序
p = 0.8
train = data[:int(len(data)*p), :] #训练
test = data[int(len(data)*p):, :] #测试
print(type(train[:,:3]))

from keras.models import Sequential #神经网络初始化
from keras.layers.core import Dense,Activation #神经网络，激活函数
net = Sequential() #建立神经网络
net.add(Dense(input_dim=3, output_dim=10)) #3个神经网络输入，下一层10个节点
net.add(Activation("relu")) #relu激活函数，中间层
net.add(Dense(input_dim=10, output_dim=1)) #10个神经网络输入，下一层10个节点
net.add(Activation("sigmoid")) #输出层
net.compile(loss="binary_crossentropy", 
            optimizer="adam") #编译模型，按照adam方法
net.fit(train[:,:3], train[:,3], nb_epoch=100, batch_size=1) #训练模型1000次
net.save_weights("sequential.net")
last = net.predict_classes(train[:,:3]).reshape(len(train))
print(last)
print(train[:,3])
print((last==train[:,3]).sum()/len(last))
```
