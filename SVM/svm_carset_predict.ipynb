```python
import pandas as pd
import numpy as np
columns = ["price", "maint", "doors", "persons", "lug_boot", "safty", "recommend"]
path = "CarDatas.txt"
data = pd.read_csv(path, header=None, sep = ' ') #这里读取要设置读取时的分隔符，默认是,分隔符
data.columns = columns
data
```

```python
data['price'].unique()
```

```python
data['safty'].unique()
```

```python
price = {'vhigh':4, 'high':3, 'med':2, 'low':1}
data["price"] = data["price"].map(price)
data["price"]
```

```python
maint = {'vhigh':4, 'high':3, 'med':2, 'low':1}
data["maint"] = data["maint"].map(maint)
data["maint"]
```

```python
data['doors'].unique()
```

```python
data['persons'].unique()
```

```python
doors = {'2':2, '3':3, '4':4, '5more':0}
persons = {'2':2, '4':4, 'more':0}
data['doors'] = data['doors'].map(doors)
data['persons'] = data['persons'].map(persons)
```

```python
# 转换5，6列
lug_boot = {'small':1, 'med':2, 'big':3} #后备箱的大小
safty = {'low':1, 'med':2, 'high':3} #安全性
data['lug_boot'] = data['lug_boot'].map(lug_boot)
data['safty'] = data['safty'].map(safty)
```

```python
from sklearn.svm import SVC #分类，SVR表示回归
svc_linear = SVC(kernel = "linear")
svc_rbf = SVC(kernel = "rbf")
svc_poly = SVC(kernel = "poly") #三种分类
```

```python
x_train = data[["price", "maint", "doors", "persons", "lug_boot", "safty"]]
y_train = data["recommend"]
svc_linear.fit(x_train, y_train)
svc_rbf.fit(x_train, y_train)
svc_poly.fit(x_train, y_train)
```

```python
svc_linear.score(x_train, y_train)
```

```python
svc_rbf.score(x_train, y_train)
```

```python
svc_poly.score(x_train, y_train)
```
