```python
import numpy as np
from sklearn.svm import SVC #分类
import matplotlib.pyplot as plt
%matplotlib inline

x_train = np.random.randn(300,2)
print(x_train)
```

```python
y_train = np.logical_xor(x_train[:,0]>0, x_train[:,1]>0)
y_train
```

```python
a = np.array([[1,2],[1,-2],[-1,-2],[-1,2]])
b = np.logical_xor(a[:,0]>0, a[:,1]>0)
print(b)
```

```python
x = np.arange(-3,3,0.01) #600个点
y = np.arange(-3,3,0.01) #600个点
xx, yy = np.meshgrid(x,y) #生成网格
xy = np.c_[xx.ravel(), yy.ravel()] #串联
xy.shape
```

```python
svc = SVC(kernel="rbf") #半径计算
svc.fit(x_train, y_train) #训练数据
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
dd = svc.decision_function(xy)
d = dd.reshape(xx.shape)
plt.imshow(d,extent=(-3,3,-3,3), cmap=plt.cm.PuOr_r)
plt.contour(xx,yy,d)
```
