```python
import sklearn.datasets as datasets
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #绘制3D
%matplotlib inline

n_sample = 1500
random_state = 170
x, y = datasets.make_blobs(n_samples=n_sample, random_state=random_state)
x.shape
```

```python
y.shape
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
y_new = KMeans(n_clusters=1).fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=y_new)
```

```python
y_new = KMeans(n_clusters=2).fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=y_new)
```

```python
y_new = KMeans(n_clusters=3).fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=y_new)
```

```python
trans = [[-0.6,0.6], [-0.4,0.8]] #手动制作数据偏差
x2 = np.dot(x, trans)
x2.shape
```

```python
plt.scatter(x2[:,0], x2[:,1], c=y)
```

```python
y_new2 = KMeans(n_clusters=3).fit_predict(x2)
plt.scatter(x2[:,0], x2[:,1], c=y_new2)
```

```python
x3, y3 = datasets.make_blobs(n_samples=n_sample, 
                             cluster_std=[1.0, 2.5, 0.5], #偏移量
                             random_state=random_state)
plt.scatter(x3[:,0], x3[:,1], c=y3)
```

```python
# kmn只能处理方差一样的情况
y3_new = KMeans(n_clusters=3).fit_predict(x3)
plt.scatter(x3[:,0], x3[:,1], c=y3_new)
```

```python
x4, y4 = datasets.make_blobs(n_samples=n_sample, 
                             cluster_std=[0.5, 0.5, 0.5], #偏移量
                             random_state=random_state)
plt.scatter(x4[:,0], x4[:,1], c=y4)
```

```python
y4_new = KMeans(n_clusters=3).fit_predict(x4)
plt.scatter(x4[:,0], x4[:,1], c=y4_new)
```

```python
x5 = np.vstack((x[y==0][:500],
                x[y==1][:100],
                x[y==2][:10]))
plt.scatter(x5[:,0], x5[:,1])
```

```python
y5 = KMeans(n_clusters=3).fit_predict(x5)
plt.scatter(x5[:,0], x5[:,1], c=y5)
```
