```python
import sklearn.datasets as datasets
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

x_train, target = datasets.make_blobs(200, centers=3)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(x_train[:,0], x_train[:,1])
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=target)
```

```python
kmn = KMeans(n_clusters=3)
kmn.fit(x_train)
```

```python
y_new = kmn.predict(x_train)
y_new
```

```python
centers = kmn.cluster_centers_
centers
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=y_new)
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=target)
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=y_new) #预测的分类
plt.scatter(centers[:,0], centers[:,1], c='r', s=100, alpha=0.4)
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=target) #默认的分类
plt.scatter(centers[:,0], centers[:,1], c='r', s=100, alpha=0.4)
```
