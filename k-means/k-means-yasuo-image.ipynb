```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
%matplotlib inline
img = plt.imread(r'./huge1.jpg')
display(img.shape, type(img))
```

```python
plt.imshow(img)
```

```python
show_img = plt.imshow(img)
```

```python
A = img.reshape((-1,3))
```

```python
A.shape
```

```python
kmn = KMeans(5)
kmn.fit(A)
B = kmn.cluster_centers_[kmn.labels_].reshape(540,540,3)
plt.imshow(B)
```

```python
plt.savefig("1.png")
```
